¿Qué motivación tenemos para utilizar algoritmos basados en la inferencia lógica en lugar de utilizar algoritmos de búsqueda para cualquier tipo de problema?

Los algoritmos basados en la inferencia lógica son útiles cuando el problema requiere razonamiento sobre conocimiento y hechos a un nivel más abstracto y flexible. A diferencia de los algoritmos de búsqueda, que exploran un espacio de soluciones de manera más exhaustiva, los algoritmos de inferencia lógica operan sobre bases de conocimiento para deducir conclusiones de manera más eficiente, utilizando reglas y relaciones lógicas. Esto permite razonar sobre hechos previamente conocidos y generar soluciones sin necesidad de recorrer explícitamente un árbol de decisiones, lo cual puede ser mucho más costoso computacionalmente en problemas complejos.

¿Qué diferencia hay entre una base de conocimientos y el ground truth?

La base de conocimientos (KB) es un repositorio de hechos, reglas y relaciones que describen un dominio de conocimiento. Incluye tanto el conocimiento general (axiomas, reglas) como hechos específicos (ground truth). El ground truth se refiere a los hechos específicos que son verificados o medidos en una instancia particular de un problema. En otras palabras, el ground truth es el conocimiento concreto y observacional dentro de un contexto específico, mientras que la base de conocimientos es más amplia y abarca tanto hechos como reglas generales del dominio.

¿Qué diferencia hay entre un valor inferido y un valor percibido?

Valor percibido es la información que proviene de los sensores o el entorno en tiempo real, como una medición directa o una observación.

Valor inferido es la conclusión o el valor que se obtiene mediante el razonamiento lógico sobre los valores percibidos y la base de conocimientos. Es un valor que no es observado directamente, sino que se deduce a partir de otros hechos.



¿Qué valores puede tomar una oración atómica de acuerdo a la lógica proposicional? ¿Cómo se modela la frase “Los valores de A comprendidos entre 0 y 9” de acuerdo a esta sintaxis?

En lógica proposicional, las oraciones atómicas solo pueden tomar dos valores: Verdadero (V) o Falso (F). La lógica proposicional es dicotómica, es decir, tiene una expresión binaria para cada proposición.

Para modelar "Los valores de A comprendidos entre 0 y 9", esta expresión no puede ser representada directamente en lógica proposicional, ya que no soporta rangos de valores. Para eso, tendrías que expresar cada valor posible de A como una proposición separada, por ejemplo:

A0 ∨ A1 ∨ A2 ∨ ... ∨ A9

Aquí, A0, A1, etc., serían proposiciones atómicas que indican si A tiene el valor correspondiente.

¿Qué algoritmos se utilizan para resolver problemas modelados con lógica proposicional?

En lógica proposicional, los algoritmos de inferencia más comunes son el encadenamiento hacia adelante y el encadenamiento hacia atrás. Estos son métodos para deducir hechos nuevos a partir de una base de conocimientos. Para representar la información, se usan cláusulas de Horn, que son disyunciones de literales en las que como máximo uno es positivo. Los algoritmos de resolución también se usan para deducir nuevas proposiciones o para probar la satisfacibilidad de un conjunto de proposiciones, como el algoritmo de resolución.

¿Qué valores puede tomar una oración atómica de acuerdo a la lógica de primer orden? ¿Cómo se modela la frase “Los valores de A comprendidos entre 0 y 9” de acuerdo a esta sintaxis?

En lógica de primer orden, las oraciones atómicas tienen una estructura más rica que en la lógica proposicional. Pueden tomar valores Verdadero (V) o Falso (F), pero la lógica de primer orden puede expresar relaciones más complejas entre objetos y sus propiedades. En lugar de ser simplemente un valor de verdad, las oraciones atómicas pueden incluir términos como objetos, funciones y predicados. Ejemplo:

Presion(tuberiaA, 100)

Esta oración indica que la presión en la tubería A es 100, y puede ser evaluada como Verdadera o Falsa.

Para modelar "Los valores de A comprendidos entre 0 y 9" en lógica de primer orden, puedes usar un predicado que indique que A tiene un valor dentro de ese rango, por ejemplo:

Entre0y9(A) ⇔ (A >= 0) ∧ (A <= 9)
	
¿Qué algoritmos se utilizan para resolver problemas modelados con lógica de primer orden?

Para lógica de primer orden, se pueden usar enfoques como el encadenamiento hacia adelante y el encadenamiento hacia atrás, similares a los de la lógica proposicional. También se utilizan algoritmos de búsqueda como A* o búsquedas de profundidad o amplitud en el espacio de estados. Sin embargo, dado que los problemas de lógica de primer orden son más complejos, los algoritmos deben ser capaces de manejar la estructura de los términos y las relaciones entre objetos.

Describa cómo los planificadores hacen uso de algoritmos de búsqueda globales para encontrar una secuencia óptima de pasos a seguir para resolver un problema
	
Los planificadores usan algoritmos de búsqueda globales como la búsqueda A* para encontrar una secuencia óptima de acciones. La búsqueda se realiza en el espacio de estados, evaluando los costos de cada transición de estado. Los planificadores definen una función heurística (por ejemplo, el número de piezas fabricadas o la cantidad de tareas completadas) para estimar la proximidad a la meta. Esta heurística ayuda a guiar la búsqueda hacia el objetivo de manera más eficiente, evitando una exploración exhaustiva de todo el espacio de estados.
